# Long Calls 

When you are betting the stock price will go up. 

```python
from Tools.options_chain import OptionsChain

ticker = "RIVN"
oc = OptionsChain(ticker, buy=False, sell=True) 
oc.set_all() # Set all class data.
calls = oc.get_calls() # Get the puts from the chain
#calls = calls[calls["sell_credit"] > 0].reset_index(drop=True)
print(calls)
```

```python
row = calls.iloc[16]
strike = row["strike"]
spread = row["strike_spread"]
expiration = row["expirationDate"]
dte = row["DTE"]
tdte = row["TDTE"]
bt = oc.backtest.get_probability(strike, option_type="put", expiration_date=expiration, return_value=False, return_dict=True)

display = f"""
===========================================================
Price: {oc.dollar_format.format(oc.stock_price)}
Strike: {oc.dollar_format.format(strike)}
Distance: {oc.percent_decimal_format.format(spread)}

----------
[Expiration]

DTE: {dte}
TDTE: {tdte}

----------
[Analysis]

Total Candles: {len(oc.candles)}
Total Periods: {bt['total']}
Total Matches: {bt['match']}
Probability: {oc.percent_decimal_format.format(bt['probability'])}

"""

print(display)
```
